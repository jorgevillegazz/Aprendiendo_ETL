#Para saber ms: ventajas de los arrays


Las listas en Python son estructuras de datos bsicas que pueden contener elementos de diferentes tipos (enteros, cadenas, otras listas, etc.). Aqu tienes un ejemplo de cmo convertir una lista en un array Numpy:

```python
import numpy as np

# crea una lista
lista = [1, 2, 3, 4, 5]

# convierte la lista en un array Numpy
array = np.array(lista)

print("Lista: ", lista)
print("Array: ", array)
```

##Regresin Lineal

Estbamos calculando la frmula o el mecanismo para encontrar la lnea que representaba este conjunto de datos, referente a los precios de los kilos de manzana, en la ciudad de Mosc. Y, por prueba y error, llegamos a un valor, que finalmente constatamos a travs de esta funcin de lgebra lineal de NumPy, para obtener la norma, que sera la pendiente. Sin embargo, podemos hacerlo de una forma ms elegante, sin la necesidad de recurrir a la prueba y error, ***aprovechando todo el potencial de NumPy como tal.***

Entonces, tenemos un valor de coeficiente ***alfa*** $$\alpha$$
 que sera la pendiente,

 un valor de ***beta***  $$\beta$$ que sera la interseccin,

 un valor ***n***$$n$$, que ser el nmero de observaciones, digamos que tenemos varios precios y varias fechas,

 el valor de ***y***$$y$$ , el precio, y el valor de ***x*** $$x$$, la fecha.

$$\alpha = pendiente$$
$$\beta = interseccion$$
$$n = observaciones$$
$$y = precio$$
$$x = fecha$$



La solucin para ***la pendiente***, cuando tenemos ***tantas observaciones***, est dada por esta ecuacin

$$\alpha = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2} $$

Ppara la interseccin, se trata de la sumatoria de los precios menos alfa multiplicado por la sumatoria de las fechas, todo dividido entre el nmero de observaciones.
 $$\beta = \frac{\sum y - \alpha \sum x}{n} $$


Entonces, la pendiente est dada por el producto entre el nmero de observaciones y la sumatoria del producto entre las fechas y los precios menos la sumatoria de las fechas multiplicada por la sumatoria de los precios, todo dividido entre el producto del nmero de observaciones por la sumatoria de cada fecha al cuadrado menos la sumatoria de todas las fechas al cuadrado.



Y obtenemos la ecuacin de la recta:

$$y = \alpha x + \beta $$

Ahora  definamos  las  vriables de las Formulas

```python
n = np.size(moscu)
y = moscu
x = fechas
```

ahora  apliquemos la formuala de la  pendiente ***a*** $$\alpha$$


$$\alpha = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2} $$

```python
a = (n*np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x**2) - np.sum(x)**2)# usamos doble asterisco ** para elevar al cuadrado
```

ahora apliquemos la formula de la imterseccin ***b*** $$\beta$$

 $$\beta = \frac{\sum y - \alpha \sum x}{n} $$

```python
b = np.mean(y)- a*np.mean(x)
```

imprimamos los resultados

```python
print(f'La pendiente es: {a}\nLa interseccin es: {b}')
```

con estos resultados  podemos aplicar ya la formula de la  ***recta*** $$y$$


```python
y = a*x + b
```
